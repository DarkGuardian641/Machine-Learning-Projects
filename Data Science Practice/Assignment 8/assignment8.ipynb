# Data Visualization I
### 1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we can find any patterns in the data.
### 2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("titanic")
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['age']=df['age'].fillna(np.mean(df['age']))
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town']=df['embark_town'].fillna(df['embark_town'].mode()[0])
df['deck']=df['deck'].fillna(df['deck'].mode()[0])
df.isnull().sum()
```

```python
sns.displot(x=df['age'], bins=10, kde=False)
```

# Distribution plot

```python
sns.jointplot(x=df['age'], y=df['fare'], kind='scatter')
```

```python
sns.jointplot(x=df['age'], y=df['fare'], kind='hex')
```

```python
sns.rugplot(x=df['fare'])
```

# Categorical plot

```python
sns.barplot(x=df['sex'], y=df['age'], data=df)
```

```python
sns.barplot(x=df['sex'], y=df['age'], data=df, estimator=np.std)
```

```python
sns.countplot(x='sex', data=df)
```

```python
sns.boxplot(x=df['age'], y=df['fare'])
```

```python
sns.boxplot(x=df['sex'], y=df['age'])
```

```python
sns.boxplot(x=df['sex'], y=df['age'], data=df, hue="survived", palette='Set2')
```

```python
sns.violinplot(x=df['sex'], y=df['age'])
```

```python
sns.violinplot(x=df['sex'], y=df['age'], data=df, hue='survived')
```

# Advanced plot

```python
sns.stripplot(x=df['sex'], y=df['age'])
```

```python
sns.stripplot(x=df['sex'], y=df['age'], data=df, jitter=False)
```

```python
sns.histplot(df['fare'], bins=20, color='blue')
plt.title('Distribution of Ticket Fare for Titanic Pasengers', fontsize=10)
plt.show()
```

```python
sns.swarmplot(x=df['sex'], y=df['age'])
```

```python
sns.swarmplot(x=df['sex'], y=df['age'], data=df, hue='survived')
```
